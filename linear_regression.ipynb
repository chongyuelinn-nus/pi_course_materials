```python
%matplotlib notebook

# import keras
# import tensorflow as tf

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# generate fake dataset

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return  t*2 +  np.random.normal(0,1,t.shape)
#     return np.exp(-t) * np.cos(2*np.pi*t)

x= np.arange(0.0, 5.0, 0.1)
# print (x)
fig = plt.figure()
y = f(x)
plt.plot(x, y, 'bo')
```

```python
# normalise
y = tf.keras.utils.normalize(y)
print (y)
```

```python
# create 1 - 10 - 10 - 1 MLP
model = tf.keras.models.Sequential() # ff mlp
model.add(tf.keras.layers.Dense(1, input_dim=1, activation=tf.nn.tanh)) # change the number of nodes
# model.add(tf.keras.layers.Dense(10, activation=tf.nn.tanh)) # change the number of nodes
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid))

optimizer = tf.keras.optimizers.RMSprop(0.01)

model.compile(loss='mean_squared_error',
            optimizer= optimizer, #'sgd',
            metrics=['mean_absolute_error', 'mean_squared_error'])
model.summary()
```

```python
model.fit(x, y, epochs=1000) # change the number of epochs
```

```python
model.fit(x.transpose(), y.transpose(), epochs=1000) # change the number of epochs
```

```python
# save model
model.save('quad_fit.model')
```

```python
# inferencing
new_model = tf.keras.models.load_model('quad_fit.model')
x_test = np.arange(0.0, 5.0, 0.1)
predictions = model.predict(x_test.tolist())
print(predictions)
```

```python
fig = plt.figure()
plt.plot(x_test, predictions, 'ro')
plt.plot(x, f(x), 'bo')
```
